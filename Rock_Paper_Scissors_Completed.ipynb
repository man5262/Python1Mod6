```python
import random as rand

def start_game():
    print("Let's play rock, paper, scissors\n")
    print("The rules are:\n\t- Rock beats scissors;\n\t- Scissors beats paper;")
    print("\t- Paper beats rock\n")

# Entering user's choice
def get_user_choice(message):
    user_choice = input(message + "? (You may press 'q' to quit the game)").lower()
    if user_choice == "" or user_choice not in message.lower().split(", ") and user_choice != 'q':
        user_choice = "wrong"
    return user_choice

# Random selection of choice number (computer's choice)
def get_computer_choice():
    return rand.randint(0, 2)  

# Get user's and computer's selections and convert them to points
def convert_selections_to_points(uchoice, cchoice):
    ret = {}
    ret["User Point"] = 0
    ret["Computer Point"] = 0
    if (uchoice != cchoice):
        if (uchoice == "rock" and cchoice == 'scissors' or
            uchoice == "paper" and cchoice == 'rock' or
            uchoice == "scissors" and cchoice == 'paper'):
            ret["User Point"] = 1
        else:
            ret["Computer Point"] = 1  
    return ret 

# Define the longest choice value
def longest_choice(choices):
    l = 0
    for item in choices:
        if len(item) > l:
            l = len(item)
    return l

message = "Rock, Paper, Scissors"
choices = message.lower().split(", ")
history = {}

# Starting game showing the game rules
start_game()
round_number = 0
round_results = {}
tmp_result = {}
while True:
    
    # Get user's and computer's choices
    user_choice = get_user_choice(message)
    if user_choice == 'q':
        break
    else:
        if user_choice != 'wrong':
    
            cc = get_computer_choice()
            computer_choice = choices[cc]
            
            # Convert selectios to awarding points and print round result
            round_score = convert_selections_to_points(user_choice, computer_choice)

            # Collecting single round data
            round_results["You selected"] = user_choice
            round_results["Computer selected"] = computer_choice
            round_results["User Point"] = round_score["User Point"]
            round_results["Computer Point"] = round_score["Computer Point"]
            if (round_score["User Point"] == 1):
                round_results["Result"] = "You win!"
            elif (round_score["Computer Point"] == 1):
                round_results["Result"] = "You lose"
            else:
                round_results["Result"] = "It's a draw"

            print(f"\nYou selected '{user_choice}'")
            print(f"Computer selected '{computer_choice}'")
            print(round_results["Result"])
            
            round_number += 1
    
            # Collecting entire game history data
            history[round_number] = {}
            for item in round_results.keys():
                history[round_number][item] = round_results[item]
   
    # Continue playing or exit?
    play_again = input("Do you want to exit the game? (enter 'q' to quit)")
    if play_again.lower() == 'q':
        
        # Print final result and history of the game before quitting the game
        user_score = 0
        computer_score = 0
        winlose = ""
        
        max_length = longest_choice(choices)
       
        for num in history.keys():
            user_score += history[num]["User Point"]
            computer_score += history[num]["Computer Point"]
        if user_score > computer_score:
            winlose = 'You win '
        elif user_score > computer_score:
            winlose = 'You loose '
        else:
            winlose = 'The draw'
        print(f"\nThe final result is: {winlose} {user_score} over {computer_score}\n")
        print("Round #\tYour Selection\tComputer Selection\tRound Result")
        for num in history.keys():
            print(str(num) + '\t' + history[num]["You selected"] + ' ' * (max_length - len(history[num]["You selected"])) + '\t' +
                  history[num]["You selected"] + ' ' * (max_length - len(history[num]["You selected"])) + '\t\t' +
                  history[num]["Result"])
        
        break
```

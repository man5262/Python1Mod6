```python
# Define MC Cafe menu
def fill_mc_menu_dict():
    return {"Coffee" : {"Espresso" : 1.50,
                        "Americano" : 2.00,
                        "Caffe latte" : 2.25,
                        "Mocha" : 2.25,
                        "Cappucino" : 2.50,
                        "Macchiato" : 2.50},
            "Tea" : {"White tea" : 1.50,
                     "Green tea" : 2.00,
                     "Oolong tea" : 2.25,
                     "Black tea" : 2.25,
                     "Earl Grey" : 2.50,
                     "Matcha" : 2.50},
            "Pastries" : {"Croissant" : 5.50,
                          "Cinnamon roll" : 4.50,
                          "Banana bread" : 4.25,
                          "Pumpkin bread" : 3.75,
                          "Sticky bun" : 4.25,
                          "Apple turnover" : 3.75}}

# Display MC Cafe menu
def show_menu(mc_menu):
    print("\t\tMC Cafe menu\n")
    for group in mc_menu.keys():
        print('\t' + group)
        for item in mc_menu[group].keys():
            print(f"\t\t{item} - ${mc_menu[group][item] :.2f}")
    
your_order = {}
cnt_orders = 0
found_item = False
found_in_orders = False
item_max_length = 0

mc_menu = fill_mc_menu_dict()

print("Welcome to MC Cafe!\n")
start_ordering = input("Would you like to see the menu? (enter 'y' - for 'YES' or any other key - for 'NO')")
if (start_ordering.lower() == 'y'):
    
    # Display the menu
    show_menu(mc_menu)

    # Take an order
    while True:
        item_ordered = input("\nPlease enter an item you want to order (enter 'q' to stop ordering):").lower()
        if (item_ordered == 'q'):
            break
        if (item_ordered == ''):
            print("You haven't entered any item!")
            continue
        else:
            for group in mc_menu.values():
                if (found_item):
                    break
                for item in group.keys():
                    if (item.lower() == item_ordered):
                        
                        found_item = True
                        
                        # Get a quantity of ordered item
                        while True:
                            err = False
                            quantity = input(f"How many {item}s would you like to order?")
                            try:
                                quantity = int(quantity)
                                if (quantity < 0):
                                    err = True
                            except:
                                    err = True
                            if (not err):
                                break
                            else:
                                print (f"Sorry, the quantity of {item} is incorrect. Please try again.")
                                
                        # If item was already ordered then just add the last entered quantity         
                        for num in range(1, cnt_orders):
                            if your_order[num]["order_item"].lower() == item.lower():
                                found_in_orders = True
                                your_order[num]["order_quantity"] += quantity
                            break
                        
                        # Add a new item info to your_order dictionary
                        if (not found_in_orders):
                            cnt_orders += 1
                            your_order[cnt_orders] = {}
                            your_order[cnt_orders]["order_item"] = item
                            your_order[cnt_orders]["order_item_price"] = group[item]
                            your_order[cnt_orders]["order_quantity"] = quantity
                        
                            # Find the longest order item name (need it for formatting purposes)
                            if len(item) > item_max_length:
                                item_max_length = len(item)
                                break
                        
                        found_in_orders = False
                        
            if (not found_item):
                print(f"Sorry, item '{item_ordered}' cannot be found in the MC Cafe menu")
            else:
                found_item = False
    
    subtotal = 0
    
    # Displaying a receipt when ordered at least one item and accumulating subtotal amount
    if (cnt_orders > 0):
        print("\n\n\t\tYour order is:\nItem\t\tPrice\tQuantity\n")
        for val in your_order.values():
            item1 = val["order_item"]
            price1 = val["order_item_price"]
            quant1 = val["order_quantity"]
            subtotal += val["order_item_price"] * val["order_quantity"]
            print(f"{item1}{' ' * (item_max_length - len(item))}\t{price1:.2f}\t{quant1}")
            
        # Receipt calculations
        tax = subtotal * .06
        print(f"=======\nSubtotal:    ${subtotal:.2f}\nTax:         ${tax:.2f}\nGrand Total: {subtotal + tax:.2f}")
        
print("\nThank you for visiting MC Cafe!")
```
